# Profit and Sales


```
new_data=['Sales','Quantity','Discount','Profit']
plt.figure(figsize=(15,10))
sns.heatmap(df[new_data].corr(),annot=True, fmt='.4f',center=0);
```

```
fig , ax = plt.subplots(figsize = (10,6))
ax.scatter(df["Sales"],df["Profit"])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show
```

# Top Cities by Sales and Profits

```
Top_sales = df.groupby("City").Sales.sum().nlargest(n =10)
Top_profits = df.groupby("State").Profit.sum().nlargest(n =10)
```

```
plt.xlabel("Cities", fontsize =15)
plt.ylabel("Total Sales",fontsize =15)
plt.title("Top Cities by Sales",fontsize =20)
plt.xticks(rotation=75)
sns.barplot(Top_sales.index, Top_sales);
```

```
plt.xlabel("State", fontsize =15)
plt.ylabel("Total Profits",fontsize =15)
plt.title("Top States by profits",fontsize =20)
plt.xticks(rotation=75)
sns.barplot(Top_profits.index, Top_profits);
```

### From the above figure it can be concluded that in cities New york City and Los Angeles have lot of Sales 
### From the above figure it can be concluded that in States New york City and California are far more profitable than the other states 

```
plt.figure(figsize = (14,6))
sns.set(font_scale=1, palette= "magma")
sns.barplot(data = df , x = "Region",y = "Profit" ,hue = "Segment")
plt.show()
```
